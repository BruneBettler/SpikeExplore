```python
import numpy as np
import json 

def exclude_channels(json_path, excluded_ids, overwrite=False):
    cleaned_data = {}
    with open(json_path, "r") as f:
        probe_dict = json.load(f)
        indices_toRemove = np.argwhere(np.isin(probe_dict["chanMap"], excluded_ids)).flatten().tolist()
        for k, v in probe_dict.items():
            if type(v) == list:
                if len(v) == len(probe_dict["chanMap"]):
                    cleaned_data[k] = np.delete(v, indices_toRemove).tolist()
            else:
                cleaned_data[k] = v

    with open(json_path, "w") as f:
        json.dump(cleaned_data, f)
        
    
json_path = r"C:\Users\social\.kilosort\probes\single_flexibleProbe_Chan3.json"
excluded_ids = [105, 146, 107, 148, 109, 150, 111, 152, 246, 154, 248, 156, 249, 157, 237, 158, 235, 159, 236, 160, 233, 162, 234, 161, 164, 231]

exclude_channels(json_path, excluded_ids)
```
